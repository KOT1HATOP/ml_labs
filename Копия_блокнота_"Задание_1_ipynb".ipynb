##Задание для гениев *
Агрегируйте числовой признак по категориальному признаку с использованием метода **без использования** библиотеки pandas, и вычислите медиану и среднее значение для каждой категории.

```python
#Агрегация

import statistics
from collections import defaultdict

diction = defaultdict(list)

df.groupby('address').agg(['median', 'mean'])
newdf = df[['address','traveltime']].dropna()

for i, j in newdf.iterrows():
    address = j[0]
    travel = j[1]

    diction[address].append(travel)

for x, c in diction.items():
    print(f'{x} - Медиана равна: {statistics.median(c)}, Среднее равно: {sum(c) / len(c)}')
```
