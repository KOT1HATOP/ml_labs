```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("/content/student-por.csv")
df
```

```python
def create_holes(dataframe, column_name, hole_percentage = 10):
    if hole_percentage < 0 or hole_percentage > 100:
        raise ValueError("Процент дыр в промежутке от 0 до 100.")

    dataframe_with_holes = dataframe.copy()
    num_holes = int(len(dataframe) * (hole_percentage / 100))
    hole_indices = np.random.choice(len(dataframe), num_holes, replace=False)
    dataframe_with_holes.loc[hole_indices, column_name] = np.nan

    return dataframe_with_holes

df1 = create_holes(df, 'age') # Добавление пропущенных значений
df2 = create_holes(df1, 'school')
df2
```

```python
#Удаление

df_cleaned = df2.dropna()
df_cleaned
```

```python
#Заполнение при помощи медианы

df_filled_median = df2.fillna(df2.median())

df_filled_median
```

```python
#Заполнение при помощи среднего

df_filled_mean = df2.fillna(df2.mean())

df_filled_mean
```

```python
#Заполнение при помощи моды

df_filled_mode = df2.fillna(df2["school"].mode()[0])

df_filled_mode
```

```python
#Заполнение при помощи интерполяции

df_interpolated = df2.interpolate()

df_interpolated
```

**Те, кто выполняет задания со * делают поиск выбросов с помощью написанного ранее класса**

Найти выбросы по числовой переменной (использовать метод основанный на среднем и на межквартильном размахе). Объясните результат

Вывести статистику с выбросами и без. Сравните и объясните результат


```python
#Методом среднего

mean = df["age"].mean()
std_deviation = df["age"].std()
threshold = 3 * std_deviation

# Идентификация выбросов
outliers = [x for x in df["age"] if abs(x - mean) > threshold]

# результат
print("Среднее значение:", mean)
print("Стандартное отклонение:", std_deviation)
print("Потенциальные выбросы:", outliers)
```

Объяснение: в данном методе выбросами стали 21 и 22, так как людей такого возраста меньше всего в датасете

```python
#При помощи межквартильного размаха

Q1 = df["age"].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3-Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

outliers2 = [x1 for x1 in df["age"] if ((x1 < lower) or (x1 > upper))]#При помощи межквартильного размаха

print("Потенциальные выбросы:", outliers2)
```

Объяснение: в данном методе выбросом стало число 22, так как людей такого возраста меньше всего в датасете

Вывод: метод основанный на среднем, посчитал что выбросами являются все люди достигшие 21 и 22 лет, а  у метода основанном на межквартильном размахе выбросами посчитал только людей в возрасте 22 лет.

###Пример реализации

```python
import pandas as pd
import numpy as np


def corr_pearson(column_1, column_2):
  cov = np.cov(column_1, column_2)
  return cov/(column_1.std() * column_2.std())
```

```python
data = {
    'Рост': [170, 155, 180, 162, 175],
    'Вес': [65, 50, 75, 55, 70]
}
df = pd.DataFrame(data)
df
```

```python
corr_pearson(df["Рост"], df["Вес"])
```
